```python
import matplotlib.pyplot as plt
import numpy as np

# Data for Without AI
time_elapsed_without_ai = {
    "Task 1": 3.15, "Task 2": 0.758, "Task 3": 0.72, "Task 4": 0.66, "Task 5": 1.057,
    "Task 6": 1.125, "Task 7": 1.057, "Task 8": 0.985, "Task 9": 1.011, "Task 10": 1.095,
    "Task 11": 1.087, "Task 12": 0.938, "Task 13": 1.011, "Task 14": 0.917, "Task 15": 1.688
}

# Data for With AI
time_elapsed_with_ai = {
    "Task 1": 0.774, "Task 2": 1.02, "Task 3": 0.712, "Task 4": 0.691, "Task 5": 0.974,
    "Task 6": 1.245, "Task 7": 1.087, "Task 8": 0.893, "Task 9": 1.11, "Task 10": 0.886,
    "Task 11": 0.817, "Task 12": 0.937, "Task 13": 0.765, "Task 14": 0.945, "Task 15": 1.081
}

# Calculate average time elapsed for each task
avg_time_without_ai = np.mean(list(time_elapsed_without_ai.values()))
avg_time_with_ai = np.mean(list(time_elapsed_with_ai.values()))

# Plotting
tasks = list(time_elapsed_without_ai.keys())
x = np.arange(len(tasks))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, list(time_elapsed_without_ai.values()), width, label='Without AI')
rects2 = ax.bar(x + width/2, list(time_elapsed_with_ai.values()), width, label='With AI')

ax.set_ylabel('Time Elapsed (seconds)')
ax.set_title('Comparison of Time Elapsed for Each Task with and without AI')
ax.set_xticks(x)
ax.set_xticklabels(tasks, rotation=45, ha='right')
ax.legend()

plt.axhline(y=avg_time_without_ai, color='b', linestyle='-', label='Avg Time Without AI: {:.2f}s'.format(avg_time_without_ai))
plt.axhline(y=avg_time_with_ai, color='r', linestyle='-', label='Avg Time With AI: {:.2f}s'.format(avg_time_with_ai))
ax.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
